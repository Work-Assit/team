```python
import re
import os
import pandas as pd
import docx2txt

def find_present_location(text, Present_Location):
    present_location = []
    highest_number = -1

    # Find the highest 4-digit number in the text
    numbers = re.findall(r'\b\d{4}\b', text)
    highest_number = max(map(int, numbers), default=-1)

    if highest_number != -1:
        for location in Present_Location:
            if location.lower() in text.lower() and str(highest_number) in text:
                present_location.append(location)

        # Find the nearest location among PVT LTD, pvt ltd., PVT LTD., Bank, and BANK
        keywords = ['PVT LTD', 'pvt ltd.','(pvt ltd.)' 'PVT LTD.', 'Bank', 'BANK','Current Address:','Current Address','current address'
                   'till date','Till date','REGION:','ltd','Location']
        nearest_location = None
        min_distance = float('inf')

        for location in present_location:
            for keyword in keywords:
                index = text.lower().find(keyword.lower())
                if index != -1:
                    distance = abs(index - text.lower().find(location.lower()))
                    if distance < min_distance:
                        min_distance = distance
                        nearest_location = location

        if nearest_location:
            present_location = [nearest_location.capitalize()]  # Capitalize the location

    return present_location[:1]

folder_path = "D:\Work-Assit\SOURCE1"  # Replace with the actual folder path
file_list = os.listdir(folder_path)

results = []
for file_name in file_list:
    if file_name.endswith(".docx"):
        file_path = os.path.join(folder_path, file_name)
        text = docx2txt.process(file_path)
        present_location = find_present_location(text, Present_Location)  # Assuming Present_Location is defined elsewhere
        results.append({"File Name": file_name, "Present Location": present_location})

df = pd.DataFrame(results)
df.to_csv("output.csv", index=False)
```
