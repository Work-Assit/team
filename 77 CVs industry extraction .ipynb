```python
import os
import re
from docx import Document
import pandas as pd

folder_path = "C:\\Users\\ask2s\\Project new CVs"
pattern = [r"\b(Banking and Finance|Insurance|banking Operation|Banking|Retail banking|investment fund|Bank pvt. ltd.|Bank ltd.|CA-firm|Teaching|Teacher|Power and Gas|Bank|Financial products|financial Products|Business banking|fashion industry|Real estate industry|Apparel Industries|broker|Account manager|Fashion Advisor|Fashion consultan|home loans|banking|Account assistant(CA firm)|Self-employee|Bank|Financial service|Banca assurance|Telecommunication|Fresher|Credit card|Electronic sector|Wealth management|B2B E-Commerce|leather Accessories|Hospital|Healthcare)\b"
]

if os.path.isdir(folder_path):
    print(f"Running folder path: {folder_path}")
    
    results = []
    
    for file_name in os.listdir(folder_path):
        if file_name.endswith(".docx"):
            file_path = os.path.join(folder_path, file_name)
            print(f"Processing file: {file_path}") 
            
            # Load from the Word document
            doc = Document(file_path)
            
            # Extract text from paragraphs and join them
            text = " ".join([p.text for p in doc.paragraphs])
            
            # Search for industries using regex pattern
            matches = []
            for p in pattern:
                match = re.search(p, text, re.IGNORECASE)
                if match:
                    matches.append(match.group())
            
            if matches:
                results.append((file_name, matches))
                print("Working sectors:", matches)
            else:
                print("No working industries found.")
            print("------------------------------")
    
    # Create a DataFrame from the results
    df = pd.DataFrame(results, columns=["File Name", "Working Sectors"])
    
    # Save the DataFrame to an Excel file
    excel_file_path = "Industry Extraction.xlsx"
    df.to_excel(excel_file_path, index=False)
    
    print(f"Results saved to: {excel_file_path}")
    
else:
    print(f"N/A: {folder_path}")
```
